#### Здесь определены вспомогательные функции

```python
def column_analysys(column):
    '''
    Функция дает общий анализ признака (столбца column):
    Считает процент пропущенных значений
    Строит гистограмму распределения
    Для численных признаков проверяет на выбросы по IQR
    '''
    # считаем количество и процент пропущенных значений
    nan_values_count = len(df.loc[:, column]) - len(df.loc[:, column].dropna())
    nan_values_perc = round(100*nan_values_count/len(df.loc[:, column]), 2)
    print(
        f'В столбце {column} пропущено {nan_values_count} значений ({nan_values_perc} %)')

    fig, ax = plt.subplots(1, 2, figsize=(16, 6))
    sns.countplot(df.loc[:, column], ax=ax[0])
    ax[0].set_title('Распределение общее')

    if df.loc[:, column].dtype != 'object':
        if check_outlies(column):
            ax[1].hist(get_IQR(column), bins=int(
                get_IQR(column).max()-get_IQR(column).min())+1)
            ax[1].set_title('Распределение без учета выбросов')
        display(pd.DataFrame(
            [df.loc[:, column].value_counts(), get_IQR(column).value_counts()]))

    return


def IQR(column):
    '''
    Функция возвращает предельные значения по IQR
    '''
    Q1 = df.loc[:, column].quantile(0.25)
    Q3 = df.loc[:, column].quantile(0.75)
    IQR = Q3 - Q1
    return (Q1 - 1.5*IQR, Q3 + 1.5*IQR)


def check_outlies(column):
    '''
    Функция проверяет, есть ли выбросы по IQR
    '''
    low, high = IQR(column)
    if df.loc[:, column].min() < low or df.loc[:, column].max() > high:
        print('Выбросы есть')
        return True
    else:
        print('Выбросов нет')
        return False


def get_IQR(column):
    '''
    Функция возвращает значения признака в столбце column,
    очищенные от выбросов по IQR
    '''
    low, high = IQR(column)
    return df[df.loc[:, column].between(low, high)].loc[:, column]


def get_stat_dif(df, column):
    '''
    Функция проверяет, есть ли статистическая разница в распределении оценок
    за экзамен по математике по признаку из столбца column с помощью теста Стьюдента
    '''
    # уникальные значения признака
    cols = df.loc[:, column].value_counts().index
    # список с неповторяющимися парами
    combinations_all = list(combinations(cols, 2))
    # уровень значимости 5% с учетом поправки Бонферони
    if len(combinations_all) > 0:
        alpha = 0.05/len(combinations_all)
    else:
        alpha = 0.05
    for comb in combinations_all:
        if ttest_ind(df[df.loc[:, column] == comb[0]].score,
                     df[df.loc[:, column] == comb[1]].score).pvalue <= alpha:
            print('Найдены статистически значимые различия для столбца', column)
            return True
            break


# словарь допустимых значений по условиям задачи, для числовых признаков зададим границы
valid_values = {'sex': ['F', 'M'],
                'age': [15, 22],
                'address': ['U', 'R'],
                'famsize': ['LE3', 'GT3'],
                'Pstatus': ['T', 'A'],
                'Medu': [0, 4],
                'Fedu': [0, 4],
                'Mjob': ['teacher', 'health', 'services', 'at_home', 'other'],
                'Fjob': ['teacher', 'health', 'services', 'at_home', 'other'],
                'reason': ['home', 'reputation', 'course', 'other'],
                'guardian': ['mother', 'father', 'other'],
                'traveltime': [0, 4],
                'studytime': [0, 4],
                'failures': [0, 3],
                'schoolsup': ['yes', 'no'],
                'famsup': ['yes', 'no'],
                'paid': ['yes', 'no'],
                'activities': ['yes', 'no'],
                'nursery': ['yes', 'no'],
                'higher': ['yes', 'no'],
                'internet': ['yes', 'no'],
                'romantic': ['yes', 'no'],
                'famrel': [0, 4],
                'freetime': [0, 5],
                'goout': [0, 5],
                'health': [0, 5],
                'score': [0, 100]
                }


def check_values(column):
    '''
    Функция проверят данные на аномальные значения 
    '''
    if column not in valid_values.keys():
        print('Для этого столбца допустимые значения не определены')
        return
    else:
        if df.loc[:, column].dtype == 'object':
            # список всех значений признака в датасете
            list1 = list(set(df.loc[:, column].dropna()))
            list2 = valid_values[column]  # список всех возможных значений
            # список невозможных значений
            list_res = list((set(list1+list2)) - set(list2))
            if len(list_res) == 0:
                print('Ошибок нет')
            else:
                print(f'Ошибки признака: {list_res}')
        else:
            low, high = valid_values[column][0], valid_values[column][1]
            if df.loc[:, column].min() < low:
                print('Ошибки признака:')
                display(df[df.loc[:, column] < low].loc[:, column])
            elif df.loc[:, column].max() > high:
                print('Ошибки признака:')
                display(df[df.loc[:, column] > high].loc[:, column])
            else:
                print('Ошибок нет')
        return


def get_boxplot(column):
    '''
    Функция строит боксплоты влияния признаков на оценку за экзамен
    '''
    fig, ax = plt.subplots(figsize=(16, 6))
    sns.boxplot(x=column, y='score', data=df, ax=ax)
    plt.xticks(rotation=45)
    ax.set_title('Boxplot for '+column)
    plt.show()
```
